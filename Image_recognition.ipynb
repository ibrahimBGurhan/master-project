```python
def loadimagesfromdir(path,fimgno,limgno):
    os.chdir(path)
    img = cv2.imread(str(fimgno)+'.jpg',0)
    flatimgs = img.flatten()
    for i in range(fimgno+1,limgno+1):
        try:
            img = cv2.imread(str(i)+'.jpg',0)
            if img is not None:
                flatimgs = np.vstack((flatimgs,img.flatten()))
        except FileNotFoundError:
            print()
    return flatimgs
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Xnorm = scaler.fit_transform(X)
Xnorm.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X)
exp_vars = pca.explained_variance_ratio_
cumvar = 0
i = 0
while cumvar <=0.95:
    cumvar = cumvar + exp_vars[i]
    i = i+1
print('The number of principal components required to explain at least 95% variation:', i)
```

```python
pca = PCA(n_components=i)
Xtrpca = pca.fit_transform(X)
```

##### Applying MultinomialNB for classification to see if the results are similar or different for each expression or class based on first research question
Assuming independence between 2304 feature attributes necessary to apply MultinomialNB.


```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

def MNBmodel(Xtrnew,Xvalnew,ytrain,yval):
    nbmodel = MultinomialNB(alpha=1)
    nbmodel.fit(Xtrnew,ytrain)
    print('Naive Bayes model parameters:\n',nbmodel.get_params())
    ypreds = nbmodel.predict(Xvalnew)
    probs = nbmodel.predict_proba(Xvalnew)
    fig, axes = plt.subplots(figsize=(10, 10))
    plot_confusion_matrix(nbmodel,Xvalnew, yval,ax=axes)
    plt.show()
    print('Classification report in test set: \n',classification_report(yval,ypreds,\
                                                               labels = ['angry','disgust','fear','happy','sad']))
    fpr, tpr, thresholds = roc_curve(yval, probs[:,0], pos_label='angry')
    plt.plot(fpr,tpr)
    fpr, tpr, thresholds = roc_curve(yval, probs[:,1], pos_label='disgust')
    plt.plot(fpr,tpr)
    fpr, tpr, thresholds = roc_curve(yval, probs[:,2], pos_label='fear')
    plt.plot(fpr,tpr)
    fpr, tpr, thresholds = roc_curve(yval, probs[:,3], pos_label='happy')
    plt.plot(fpr,tpr)
    fpr, tpr, thresholds = roc_curve(yval, probs[:,4], pos_label='sad')
    plt.plot(fpr,tpr)
    plt.grid()
    plt.title('AUC under ROC of all classes as One-vs-rest:'+str(roc_auc_score(yval, probs, multi_class='ovr')))
    plt.legend(['angry','disgust','fear','happy','sad'])
    plt.show()
    return
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xnorm, y, test_size=0.3, random_state=1)
MNBmodel(X_train, X_test, y_train, y_test)
```

It can be seen that the precision is almost same for angry, fear and sad by very different from happy and disgust. Thus, alternative hypothesis is accepted for the 2nd research question.
